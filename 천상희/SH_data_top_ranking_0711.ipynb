#### 4. Top Ranking 합치기

**Merge multiple dataframes** 나라별 랭킹 데이터 합치기

```python
import pandas as pd
from functools import reduce

top_merged = [top_ww, top_ko, top_hk, top_jp, top_th, top_us]
df_top = reduce(lambda left, right: pd.merge(left,right,on=['Title'], how='outer'), top_merged)
```

**컬럼명 변경**

```python
df_top = df_top.rename(columns={"ï»¿Ranking_WW": "Ranking_WW", "ï»¿Ranking_KO": "Ranking_KO", "ï»¿Ranking_HK": "Ranking_HK", "ï»¿Ranking_Jp": "Ranking_JP", "ï»¿Ranking_TH": "Ranking_TH", "ï»¿Ranking_US": "Ranking_US"})
df_top
```

```python
df_top = df_top.rename(columns=str.lower)
df_top.head(2)
```

**특수 문자 제거**

```python
df_top['points_worldwide'] = df_top['points_worldwide'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df_top['points_ko'] = df_top['points_ko'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df_top['title'] = df_top['title'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df_top['title'] = df_top['title'].str.replace(pat=r'Â', repl=r'', regex=True)
df_top
```

**중복되는 영화 제거**

```python
df_top = df_top.drop_duplicates(['ranking_ww'], keep = 'first')
df_top
```

```python
df_top.reset_index(drop=True, inplace=True)
df_top
```

```python
df_top = df_top.iloc[:101]
df_top
```

**Null값을 0으로 대체**

```python
df_top= df_top.fillna(0)
df_top
```

**int로 변경**

```python
df_top["points_ko"] = df_top["points_ko"].astype(object).astype(int)
df_top.info()
```

```python
df_top['ranking_ww'] = df_top['ranking_ww'].astype('int')
df_top['ranking_ko'] = df_top['ranking_ko'].astype('int')
df_top['ranking_hk'] = df_top['ranking_hk'].astype('int')
df_top['points_hk'] = df_top['points_hk'].astype('int')
df_top['ranking_jp'] = df_top['ranking_jp'].astype('int')
df_top['points_jp'] = df_top['points_jp'].astype('int')
df_top['ranking_th'] = df_top['ranking_th'].astype('int')
df_top['points_th'] = df_top['points_th'].astype('int')
df_top['ranking_us'] = df_top['ranking_us'].astype('int')
df_top['points_us'] = df_top['points_us'].astype('int')
df_top.info()
```

**title에서 소문자 및 특수문자 제거**
